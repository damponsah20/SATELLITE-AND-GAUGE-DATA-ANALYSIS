PLOTTING GRAPH FOR THE YEAR 2000


```python
import matplotlib.pyplot as plt
```

```python
gauge_data = year_2000['GAUGE DATA ']
satellite_data=year_2000['SATELLITE DATA']
MONTHS = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.plot(MONTHS,gauge_data,color='r',marker='o',label='gauge data');

plt.xlabel('MONTHS')
plt.ylabel('gauge data');
plt.legend()
plt.twinx()
plt.plot(MONTHS,satellite_data,color='b',marker='o',label='satellite data');
plt.ylabel('satellite data')
plt.legend()
plt.title('A GRAPH OF MONTHLY GAUGE AND SATELLITE DATA FOR THE YEAR 2000');
```
